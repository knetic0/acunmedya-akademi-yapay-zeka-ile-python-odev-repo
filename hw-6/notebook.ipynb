```python
# Pandas kutuphanesini pd alias ile import ediyorum
import pandas as pd
# grafik senaryolari icin matplotlib kütüphanesini import ediyorum
import matplotlib.pyplot as plt
# numpy kütüphanesini import ediyorum
import numpy as np
# seaborn kütüphanesini import ediyorum
import seaborn as sns
```

```python
# HousingData.csv dosyasını okuyor ve df isimli bir DataFrame'e atıyorum
df = pd.read_csv("HousingData.csv")
```

```python
# isnull metodu ile eksik verileri kontrol ediyorum
# Bu metot, her sütundaki eksik değerlerin sayısını döndürür
df.isnull().sum()
```

```python
# fillna metodu ile eksik verileri dolduruyorum
# df.mean() ile eksik değerleri sütunun ortalaması ile dolduruyorum
# inplace=True parametresi ile değişiklikleri doğrudan df üzerinde yapıyorum
df.fillna(df.mean(), inplace=True)

# yeniden isnull metodu ile eksik verileri kontrol ediyorum
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# to_csv metodu ile temizlenmiş veriyi yeni bir CSV dosyasına kaydediyorum
df.to_csv("HousingData_Cleaned.csv", index=False)
```

```python
plt.figure()
plt.scatter(df['LSTAT'], df['MEDV'], alpha=0.5)
# Basit doğrusal regresyon eğrisi hesaplama
# numpy kütüphanesinin polyfit fonksiyonunu kullanarak LSTAT ve MEDV arasındaki ilişkiyi modelliyorum
# polyfit fonksiyonu, birinci dereceden (doğrusal) bir polinom fit eder
# slope ve intercept değerlerini alıyorum
slope_lstat, intercept_lstat = np.polyfit(df['LSTAT'], df['MEDV'], 1)
# slope = eğimin değeri
# intercept = y eksenini kestiği nokta
# fit edilen doğrusal fonksiyonu çiziyorum
# x eksenindeki değerleri belirliyorum
# 100 eşit aralıklı değer oluşturuyorum
# x eksenindeki değerleri kullanarak y eksenindeki değerleri hesaplıyorum
# y eksenindeki değerleri çiziyorum
x_vals_lstat = np.linspace(df['LSTAT'].min(), df['LSTAT'].max(), 100)
y_vals_lstat = slope_lstat * x_vals_lstat + intercept_lstat
plt.plot(x_vals_lstat, y_vals_lstat, color='red')
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.title('LSTAT ile MEDV İlişkisi')
plt.show()
```

```python
plt.figure()
plt.scatter(df['AGE'], df['MEDV'], alpha=0.5)
# Basit doğrusal regresyon eğrisi hesaplama
slope_age, intercept_age = np.polyfit(df['AGE'], df['MEDV'], 1) # x ve y verilerine en iyi uyan polinomu (burada derece 1 yani doğrusal) bulur.
# x = AGE, y = MEDV
# return olarak slope ve intercept döner
# slope = eğim, intercept = kesişim noktası
# x_vals_age ve y_vals_age ile regresyon doğrusunu çiziyorum
# x_vals_age, AGE sütununun min ve max değerleri arasında 100 eşit aralıklı değer döndürür
# y_vals_age ise bu x değerlerine karşılık gelen MEDV değerlerini döndürür
# Bu sayede regresyon doğrusunu çizebiliyorum
x_vals_age = np.linspace(df['AGE'].min(), df['AGE'].max(), 100)
y_vals_age = slope_age * x_vals_age + intercept_age
plt.plot(x_vals_age, y_vals_age, color='red')
plt.xlabel('AGE')
plt.ylabel('MEDV')
plt.title('AGE ile MEDV İlişkisi')
plt.show()
```
